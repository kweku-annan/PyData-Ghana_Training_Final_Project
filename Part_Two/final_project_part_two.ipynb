## Convid Information on June 16, 2021

```python
# Dropping duplicates in dataset
data.drop_duplicates(inplace=True)
data.head()
```

```python
data.shape
```

```python
# Getting information for June 16, 2021
june_16_2021 = data.loc[data["date"] == "2021-06-16"]
june_16_2021.set_index("date", inplace=True)
june_16_2021
```

#### State Name

```python
state_name = june_16_2021["state"][0]
# state_name = june_16_2021[1]
state_name
```

#### Today's Date

```python
today = june_16_2021.index[0]
```

#### Confirmed or Suspected Cases

```python
june_16_2021.columns
```

```python
cumulative_cases = float(june_16_2021["actuals.cases"][0]) + float(june_16_2021["actuals.newCases"])
cumulative_cases
```

#### Cumulative Deaths

```python
cumulative_deaths = float(june_16_2021["actuals.deaths"]) + float(june_16_2021["actuals.newDeaths"])
cumulative_deaths
```

#### New Cases

```python
new_cases = float(june_16_2021["actuals.newCases"])
new_cases
```

#### New Deaths

```python
new_deaths = float(june_16_2021["actuals.newDeaths"])
new_deaths
```

#### Cases per 100k Population


```python
cases_per_100k = june_16_2021["metrics.testPositivityRatio"][0]
cases_per_100k
```

#### Ratio of Population that has Completed Vaccination
    

```python
completed_vaccination = round(float(june_16_2021["metrics.vaccinationsCompletedRatio"]), 2)
completed_vaccination
```

#### Ratio of Population that has Initiated Vaccination

```python
initiated_vaccination = round(float(june_16_2021["metrics.vaccinationsInitiatedRatio"]), 2)
initiated_vaccination
```

### Data For Dashboard

```python
dashboard_data = {
    "State": state_name,
    "Date": today,
    "New cases": new_cases,
    "New deaths": new_deaths,
    "Cumulative cases": cumulative_cases,
    "Cumulative deaths": cumulative_deaths,
    "Completed Vaccination": completed_vaccination,
    "Initiated Vaccination": initiated_vaccination,
}
dashboard_data
```

## Relationship between Actual Covid Cases and Vaccinations Completed.

```python
import seaborn as sns
```

```python
sns.lineplot(data=data, x="actuals.cases", y="metrics.vaccinationsCompletedRatio")
```

From the graph above, it can be seen that as the number of cases rises, the ratio of people getting vaccinated also rises. This means that there is a positive relationship between the number of actual cases and the ratio at which people get vaccinated.
